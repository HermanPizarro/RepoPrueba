# Medición 1: Medidas de tendencia central.

Describir un comportamiento general de los datos y particularmente describen su centro.

- Media: Promedio aritmético el cual determine el número que se requiere para realizar una distribución equitativa entre los datos.
    - Problema de la media: Sensible a los valores extremos.

- Mediana: Centro de los datos, el destaca el punto en donde se contiene al 50% de los datos mas bajos o mas altos.

- Moda: Representa al dato con mayor frecuencia.

## Visualización de medidas descriptivas

La mejor alternativa que tenemos para visualizar medidas descriptivas es el histograma.

```python
import seaborn as sns # Libreria especialista en gráficos estadísticos
```

```python
# Crear un histograma
sns.histplot(data=df1, x = "Peso_Total")
```

```python
data = pd.read_excel("mpg.xlsx")
data.head(30)
```

```python
df1 = data[["mpg", "displacement", "horsepower", "weight", "acceleration", "origin"]]
df1.head() # La extensión .head() nos permite imprimir las cinco primeras filas.
```

```python
df1.describe()
```

*Elementos del describe:*

- count: Cantidad de elemenos NO NULOS que tiene la variable.
- mean: Promedio o media aritmética.
- std: Desviación estándar.
- min: Valor mínimo.
- 25%: Primer Cuartil.
- 50%: Mediana o Segundo Cuartil.
- 75%: Tercer Cuartil.
- max: Valor máximo.

```python
# Crear un histograma donde la media es mayor que la mediana.
sns.histplot(data=df1, x = "displacement", kde=True)
```

```python
# Crear un histograma donde la media es similar a la mediana
sns.histplot(data=df1, x = "acceleration", kde=True)
```
